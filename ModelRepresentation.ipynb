```python
import numpy as np
import matplotlib.pyplot as plt 
plt.style.use('./deeplearning.mplstyle')

#Shembull i trajnimeve 
x_train = np.array([2104.0, 1416.0, 852.0])
y_train = np.array([400.0, 232.0, 178.0])
x_train = np.array([214.0, 231.0])
y_train = np.array([12.0, 123.0])
print(f" x_train = {x_train}")
print(f" y_train = {y_train}")

#Tash numrin e training example -m
print(f"x_train.shape: {x_train.shape}")
m=x_train.shape[0]
print(f"Number of training example: {m}")

#ose 
m=len(x_train)
print(f"Number of training example is: {m}")

m=len(y_train)
print(f"Number 0f training example is: {m}")

#Training example of x_i dhe y_i to denote i

i=1 # qikjo e rrit modelin apo e trajnon per 1, sa here te bajm run rritet per 1 nese ish ka 0 se kish rrit hiq

x_i=x_train[i]
y_i=y_train[i]
print(f"(x`({i}), y`({i}))=({x_i}, {y_i})")

i=0 
x_i=x_train[i]
y_i=y_train[i]
print(f"(x`({i}), y`({i}))=({x_i}, {y_i})")

#Tash kemi me i plot datas duke perdorur scatter() 

plt.scatter(x_train, y_train, marker="x", c="r") # marker edhe c i shfaqin red crosses( edhe bydefault dalin edhe pikat e kalterta)
plt.scatter(x_train, y_train, marker="x", c="y")

plt.title("Housing Prices")
plt.xlabel("Prices")
plt.ylabel("Size")
plt.show()

#Tash e kemi model function per f qe i ka parametrat e w dhe b
# 1 shembull te thjeshte veq sa per me i printu keta parametra 
w=100
b=100
print(f"w:{w}")
print(f"b:{b}")

#Now kemi me i definu nje funksion qe me gjet vleren e funk f e w,b te x prej i dmth f_wb
def compute_model_output(x, w, b):

    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb

#Tash kemi me thirr compute_model_output dhe kemi me plot the data

tmp_f_wb=compute_model_output(x_train, w, b)
#Plot the model prediction
plt.plot(x_train, tmp_f_wb, c="b", label="Our Predication")
#Plot the data points
plt.scatter(x_train, y_train, marker="x", c="r", label="Actual Values")

plt.title("Housing Prices")
plt.xlabel("Prices")
plt.ylabel("Size")
plt.legend()
plt.show

#Tash qe e kemi modelin mujm me ba original prediction, psh

w=10
b=12
x_i=1.2
cost_1200sqft= w*x_i+b

print(f"${cost_1200sqft:.0f} thousand dollars")

```
